```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('Lligues europees UEFA.csv',encoding='iso-8859-1', sep = ';')
df.info()
df.describe()
```

```python
# Seleccionar dos columnas para comparar
column1 = "xG"
column2 = "GA"

# Calcular el p-valor para las dos columnas
alpha = 0.05
t_statistic, p_value = ttest_ind(df[column1], df[column2])
if p_value < alpha:
    print(f"{column1} vs {column2}: P-value = {p_value:.4f} (reject null hypothesis)")
else:
    print(f"{column1} vs {column2}: P-value = {p_value:.4f} (cannot reject null hypothesis)")
```
