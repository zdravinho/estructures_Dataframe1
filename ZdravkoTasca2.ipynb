```python
import pandas as pd
from sklearn.model_selection import train_test_split

columnas = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
df = pd.read_csv("housing data.csv",names = columnas)

df.head()
```

```python
# Assign column names to the DataFrame



train, test = train_test_split(df, test_size=0.2, random_state=42)

# Analizar el conjunto de trenes
print("Train set:")
print(train.describe())

# Analizar el conjunto de prueba
print("\nTest set:")
print(test.describe())
```

```python
from sklearn.preprocessing import StandardScaler

# Selecciona solo las variables numéricas que deseas estandarizar
num_vars = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df_num = df[num_vars]

# Crea un objeto StandardScaler y ajústalo a los datos numéricos
scaler = StandardScaler()
scaler.fit(df_num)

# Transforma los datos numéricos estandarizándolos
df_num_std = scaler.transform(df_num)
df_num_std
# Reemplaza los valores originales en el DataFrame con los valores estandarizados
df[num_vars] = df_num_std
df.head()
```

# 3 ejercicio

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Resumen estadístico
print(df["CRIM"].describe())

# Gráfico de densidad
sns.displot(df["CRIM"], kde=True)
plt.show()
```

```python
# Resumen de frecuencia
print(df["CHAS"].value_counts())

# Gráfico de barras
sns.countplot(x="CHAS", data=df)
plt.show()
```

```python
# Resumen estadístico
df["RM^2"] = df["RM"]**2
print(df[["RM", "RM^2"]].corr())

# Gráfico de dispersión
sns.scatterplot(x="RM", y="RM^2", data=df)
plt.show()
```
